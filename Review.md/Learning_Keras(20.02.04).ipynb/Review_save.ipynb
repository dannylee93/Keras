```python
import numpy as np
import pandas as pd

df1 = pd.read_csv('kospi200.csv', index_col=0, header=0, encoding='cp949', sep=',')
df2 = pd.read_csv('samsung.csv', index_col=0, header=0, encoding='cp949', sep=',')


# kospi200의 모든 데이터
for i in range(len(df1.index)):                        # 거래량 str에서 int로 변경
    df1.iloc[i,4] = int(df1.iloc[i,4].replace(',',''))

# 삼성전자의 모든 데이터
for i in range(len(df2.index)):                        # 거래량 str에서 int로 변경
    for j in range(len(df2.iloc[i])):
        df2.iloc[i,j] = int(df2.iloc[i,j].replace(',',''))
        
print(df1)
print(df1.shape)

df1 = df1.sort_values(by=['일자'], ascending=[True])
df2 = df2.sort_values(by=['일자'], ascending=[True])

df1 = df1.values
df2 = df2.values

np.save('kospi200.npy', arr=df1)
np.save('samsung.npy', arr=df2)
```
